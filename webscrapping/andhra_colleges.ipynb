```python
from urllib.request import urlopen as ureq
from bs4 import BeautifulSoup as soup
import requests
import mechanize
import time
```

```python
def clean_info(string):
    tmp = string.split()
    clean_string = ''.join(tmp)
    t3=clean_string
    return(t3)
```

```python
mainpage_url = "https://collegedunia.com/engineering/andhra-pradesh-colleges"
#base_url='https://cee.kerala.gov.in/collegelist/main/'
#attrib=["engg_G","engg_N","engg_S"]
```

```python
u_client= ureq(mainpage_url)
mainpage_html=u_client.read()
u_client.close()
mainpage_soup=soup(mainpage_html, 'html.parser')
```

```python
college_links=mainpage_soup.findAll('div',{'class':'clg-name-address'})
```

```python
len(college_links)
```

```python
college_data=[]
links=[]
for link in college_links:
    row_data=[]
    row_data.append(link.a.text)
    college_url=link.a['href']
    
    u_client= ureq(college_url)
    collegepage_html=u_client.read()
    u_client.close()
    collegepage_soup=soup(collegepage_html, 'html.parser')
    
    row_data.append(clean_info(collegepage_soup.findAll('p',{'class':'lr_detail'})[0].text))
    
    college_data.append(row_data)
```

```python
len(college_data)
```

```python
import csv

with open("andra_colleges.csv", "w") as f:
    writer = csv.writer(f)
    writer.writerows(college_data)
```

```python
driver.get(mainpage_url)
#driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")

SCROLL_PAUSE_TIME = 5

# Get scroll height
last_height = driver.execute_script("return document.body.scrollHeight")

i=0
while True:
    # Scroll down to bottom
    driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
    i+=1

    # Wait to load page
    time.sleep(SCROLL_PAUSE_TIME)

    # Calculate new scroll height and compare with last scroll height
    new_height = driver.execute_script("return document.body.scrollHeight")
    if new_height == last_height:
        break
    last_height = new_height

print(i)
```
