```python
def clean_info(string):
    tmp = string.split()
    clean_string = ''.join(tmp)
    t3=clean_string
    return(t3[t3.find(':')+1:])
```

```python
mainpage_url = "https://cee.kerala.gov.in/collegelist/main/index.php"
base_url='https://cee.kerala.gov.in/collegelist/main/'
attrib=["engg_G","engg_N","engg_S"]
```

```python
u_client= ureq(mainpage_url)
mainpage_html=u_client.read()
u_client.close()
mainpage_soup=soup(mainpage_html, 'html.parser')
```

```python
college_data=[]
for category in attrib:
    container=mainpage_soup.findAll('div',{"id":category})
    length=len(container[0].table.findAll('tr'))
    
    for index in range(1,length):
        row=[]
        if(index<10):
            row.append(clean_info(container[0].table.findAll('tr')[index].text[1:-12]))
        else:
            row.append(clean_info(container[0].table.findAll('tr')[index].text[2:-12]))
        row.append(container[0].table.findAll('tr')[index].a['href'])
        row.append(base_url+container[0].table.findAll('tr')[index].a['href'])
        
        url=base_url+container[0].table.findAll('tr')[index].a['href']
        
        driver = webdriver.Chrome('C:/Users/hp/Downloads/Compressed/chromedriver_win32_4/chromedriver.exe',chrome_options=chrome_options)
        driver.get(url)
        infopage_html=driver.page_source
        infopage_soup=soup(infopage_html,'lxml')
        #print("Chrome Browser Initialized in Headless Mode")
        driver.quit()
        print("Driver Exited")
        row.append(clean_info(infopage_soup.findAll('div',{'class':'row'})[-4].text))
        row.append(clean_info(infopage_soup.findAll('div',{'class':'row'})[-3].text))
        row.append(clean_info(infopage_soup.findAll('div',{'class':'row'})[-2].text))

        
        #url=base_url+container[0].table.findAll('tr')[index].a['href']
        college_data.append(row)
        print(len(college_data))
```
